```python
def compute_pay(hours, rate):
    if hours <= 40:
        pay = hours * rate
    else:
        regular_pay = 40 * rate
        overtime_pay = (hours - 40) * (1.5 * rate)
        pay = regular_pay + overtime_pay
    return pay

hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

pay = compute_pay(hours, rate)
print("Pay:", pay)
```

```python
def compute_grade(score):
    if 97 <= score <= 100:
        return '1.0'
    elif 93 <= score < 97:
        return '1.25'
    elif 89 <= score < 93:
        return '1.5'
    elif 85 <= score < 89:
        return '1.75'
    elif 81 <= score < 85:
        return '2.0'
    elif 77 <= score < 81:
        return '2.25'
    elif 74 <= score < 77:
        return '2.5'
    elif 0 <= score < 74:
        return '3.0'
    else:
        return 'Invalid Score'


score = float(input("Enter score: "))
grade = compute_grade(score)
print("Grade:", grade)
```
