```python
total = 0
count = 0

while True:
    try:
        num_input = input("Enter a number: ")
        if num_input.lower() == 'done':
            break
        num = float(num_input)
        total += num
        count += 1
    except ValueError:
        print("Invalid input")

if count == 0:
    print("No numbers entered.")
else:
    average = total / count
    print(total, count, average)
```
