```python
def count_words():
    line = input('Please enter a line of text: ')
    word_list = line.split()
    word_count = {}

    for word in word_list:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    for word, count in word_count.items():
        print(f'Word: {word}, Count: {count}')

count_words()
```
